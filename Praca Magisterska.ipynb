```python
column = df.DepTime
date = df.FlightDate


def date_hour(date, column):
    column = [int(x) for x in column]
    for i in range(0, len(column)):
        if column[i] < 10:
            column[i] = "00:0" + str(column[i])
        elif column[i] < 60:
            column[i] = "00:" + str(column[i])
        elif column[i] >= 2400:
            column[i] = "00:00"
        else:
            column[i] = str(column[i])[:-2] + ":" + str(column[i])[-2:]
    xd = [str(i) for i in column]
    date = [str(i)[:-8] for i in date]
    res = [i + j for i, j in zip(date, xd)]
    datetime = pd.to_datetime(res)
    return res


df["DepDateTime"] = date_hour(df.FlightDate, df.DepTime)
```
